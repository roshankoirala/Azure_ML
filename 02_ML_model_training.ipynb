```python
from azureml.core import Workspace, Experiment, Environment, ScriptRunConfig
from azureml.core.conda_dependencies import CondaDependencies
```

# Run a model from script 

Here we train a machine learning model in `scikit-learn` package and register the outcome. The process is similar to the previous notebook. 

```python
ws = Workspace.from_config()
```

```python
myenv = Environment(name='environemnt-002')
myenv_dep = CondaDependencies.create(conda_packages=['scikit-learn', 'pandas', 'pip'],
                                    pip_packages=['azureml-defaults'])
myenv.python.conda_dependencies = myenv_dep
myenv.register(ws) # optional step 
```

```python
script_config = ScriptRunConfig(source_directory='.', 
                               script='experiment_2.py', 
                               environment=myenv)
```

```python
experiment = Experiment(workspace=ws, name='exp-003')
run = experiment.submit(config=script_config)
run.wait_for_completion(show_output=True)
```
