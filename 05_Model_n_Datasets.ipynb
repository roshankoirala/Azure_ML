```python
from azureml.core import Workspace, Environment, ScriptRunConfig, Experiment
from azureml.core.conda_dependencies import CondaDependencies
```

# Model from registered data 

In this notebook we retrive a registered data in a python script. We provide name of the dataset as an argument of the `ScriptRunConfig` object. The script parse the dataset name. In this way we don't have to modify the script if we change the dataset in the model. 

Once we retrive the data we can do all the works based on it. In the script we are minimalistic with the data preprocessing and calculate the length of the data. 

```python
ws = Workspace.from_config()
```

```python
env = Environment('myenv')
packages = CondaDependencies.create(conda_packages=['pip'], 
                                   pip_packages=['azureml-defaults', 
                                                 'azureml-dataprep[pandas]'])
env.python.conda_dependencies = packages 

script_config = ScriptRunConfig(source_directory='.', 
                               script='experiment_5.py', 
                               arguments=['--ds', 'Irish Data'], 
                               environment=env)
```

```python
experiment = Experiment(workspace=ws, name='exp-005')
run = experiment.submit(config=script_config)
run.wait_for_completion(show_output=True)
```
